```python
#removing the non_numeric record and hence displaying if still exists !
df_copy=df.copy()
df_copy=df_copy.drop(df_copy.index[10472])
df_copy[~df_copy['Reviews'].str.isnumeric()]
```

```python
df_copy['Reviews']=df_copy['Reviews'].astype(int)
df_copy.info()
```

```python
df['Size'].unique()
```

```python
df_copy['Size']=df_copy['Size'].str.replace('M',"000")
df_copy['Size']=df_copy['Size'].str.replace("+",'')
df_copy['Size']=df_copy['Size'].str.replace("k",'')
df_copy['Size']=df_copy['Size'].replace("Varies with device",np.nan)
df_copy['Size'].unique()
```

```python
df_copy["Size"]=df_copy["Size"].astype(float)
df_copy.info()
```

```python
df_copy.rename(columns={'Installs':'Mininstalls'},inplace='true')
```

```python
df_copy['minInstalls']=df_copy['minInstalls'].str.replace('+','')
df_copy['minInstalls']=df_copy['minInstalls'].str.replace(',','')
df_copy['minInstalls']=df_copy['minInstalls'].astype(float)
```

```python
df_copy['minInstalls'].info()
```

```python
df_copy.head()
```

```python
#handling price column 
df_copy['Price']=df_copy['Price'].str.replace("$",'')
df_copy['Price']=df_copy['Price'].astype(float)
```

```python
df_copy['Price'].info()
```

```python
#handling date and time values correctly 

df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
```

```python
df_copy.info()
```

```python
df_copy.to_csv('cleaned_playstoredataset.csv')
print('='*130)
```

# TOP 10 APP CATEGORIES

```python
plt.figure(figsize=(15,6))
sns.barplot(x=category.index[:10], y ='count',data = category[:10],palette='hls')
plt.title('Top 10 App categories')
plt.xticks(rotation=90)
plt.show()
```

1. Free apps receive significantly more installs than paid apps

# TOP 5 MOST INSTALLED APPS IN EACH POPULAR CATEGORY 

```python
dfa = df_copy.groupby(['Category' ,'App'])['Mininstalls'].sum().reset_index()
dfa = dfa.sort_values('Mininstalls', ascending = False)
apps = ['GAME', 'COMMUNICATION', 'PRODUCTIVITY', 'SOCIAL' ]
sns.set_context("poster")
sns.set_style("darkgrid")

plt.figure(figsize=(40,30))

for i,app in enumerate(apps):
    df2 = dfa[dfa.Category == app]
    df3 = df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data= df3,x= 'Mininstalls' ,y='App' )
    plt.xlabel('Installation in Millions')
    plt.ylabel('')
    plt.title(app,size = 20)
    
plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
```

## INSIGHTS
- Most popular game is Subway Surfers.
- Most popular communication app is Hangouts.
- Most popular productivity app is Google Drive.
- Most popular social app is Instagram.
